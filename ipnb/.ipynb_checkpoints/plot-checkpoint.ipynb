```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline
x = []
y = []
count = 0
with open('../errorFunc.csv') as f:
    reader = csv.reader(f)
    l = [row for row in reader]
for row in l:
    count += 1
    x.append(int(row[0].split()[0]))
    y.append(float(row[0].split()[1]))
plt.plot(x,y)
```
